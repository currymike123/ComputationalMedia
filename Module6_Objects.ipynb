#### Class

Classes have four elements: name, data, constructor, and methods.

<ol>
<li><bold>Class Name:</bold>  The name is specified by "class NameOfClass".  Then enclose the class in curly braces.  Class names are commonly captialized.</li>
<li><bold>Data:</bold>  The data for the class is the collections of it's variables.</li> 
<li><bold>Constructor:</bold>  The constuctor is a speical function that is run when an object is created.  It is where you give instructions on how to setup the object.  It always has the same name as the class. </li> 
<li><bold>Methods:</bold>  Methods are functions that exist inside the class.  The are written the same way as normal functions with name, arguements, and return type.</li> 
</ol>

````
//The class name
class Ball {

//The data

float x;
float y;
float xSpeed;
float ySpeed;

//The constructor
Ball(){
    
    x = 100;
    y = 100;
    xSpeed = 1;
    ySpeed = 3;

}

//Methods

void speed(){

    x = x + xSpeed;
    y = y + ySpeed;

}

void bounce(){

    if((x > width) || (x < 0>)){
        xSpeed = xSpeed * -1;
    }

    if ((y > height) || (y < 0)){
        ySpeed = ySpeed * -1;
    }

}

void display(){

    stroke(0);
    fill(175);
    ellipse(x,y,15,15);
}

}
````
